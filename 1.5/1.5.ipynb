```python
import pandas as pd
df = pd.read_csv(".\\ds_salaries.csv",index_col=0)
```

Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.) 

```python
df.groupby("job_title").count()
```

Какая средняя и медианная зарплата по группам вакансий?

```python
df.groupby("job_title")["salary_in_usd"].agg(["mean","median"])
```

Какая средняя и медианная зарплата по каждому региону?

```python
df.groupby("company_location")["salary_in_usd"].agg(["mean","median"])
```

Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
df.groupby("job_title")["salary_in_usd"].agg(["mean"]).sort_values("mean",ascending=False).head(1)
```

Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
group_location = df.groupby('company_location')['salary_in_usd'].count().reset_index()
group_location["percent"] = group_location['salary_in_usd']*100 / group_location.salary_in_usd.sum()
group_location
```

Какая корреляция уровня опыта от зарплаты?

```python
df["cat_dig"] = pd.Categorical(df.experience_level)
df['code'] = df["cat_dig"].cat.codes
df["code"].corr(df["salary_in_usd"])
```

Сколько должностей в наборе данных?

```python
print("Число должностей",len(df.groupby("job_title").groups))
```

Какие 10 наиболее часто встречающихся должностей?

```python
df.groupby("job_title")["salary"].count().reset_index().sort_values(by="salary", ascending=False).rename(columns={"salary":"total_count"}).head(10)
```
